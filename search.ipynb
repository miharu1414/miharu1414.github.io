```python
import requests
from bs4 import BeautifulSoup
# Webページを取得して解析する

load_url = "https://www.ski-gelende.com/gelende/"
html = requests.get(load_url)
soup = BeautifulSoup(html.content, "html.parser")

# HTML全体を表示する
print(soup)
```

```python
import re
html = str(soup)
urls = []
names = []
cnt = 3
while True:
    try:
        url = re.search( r'(?<=<li><a href=").+?(?=">)', html).group(0)
        urls.append(url)
        
        index = html.find("<li><a href=\""+url+"\">")
        html = html[index:]
        html = html.replace('<li><a href=\"'+url+'\"','')
        
        name = re.search( r'(?<=>).+?(?=</a>)', html).group(0)
        names.append(name)
        if name == "箱館山スキー場":
            break
    except:
        
        break
print(names,urls)
print(len(names))
html
```

```python
import re
html = str(soup)

info = []

for i in urls:
  try:
    load_url = i
    html = requests.get(load_url)
    soup = BeautifulSoup(html.content, "html.parser")
    html = str(soup)
    info_sabu = []
    height = re.search(r'(?<=トップ標高：</dt><dd class="col-sm-7 col-xs-6 p5">).+?(?=<)', html).group(0)
    info_sabu.append(height)
    index = html.find("</dt><dd class=\"col-sm-7 col-xs-6 p5\">"+height+"</dd>")
    html = html[index:]
    html = html.replace("</dt><dd class=\"col-sm-7 col-xs-6 p5\">"+height+"</dd>",'')
    syado = re.search(r'(?<=最大斜度：</dt><dd class="col-sm-7 col-xs-6 p5">).+?(?=<)', html).group(0)
    info_sabu.append(syado)
    index = html.find("</dt><dd class=\"col-sm-7 col-xs-6 p5\">"+syado+"</dd>")
    html = html[index:]
    html = html.replace("</dt><dd class=\"col-sm-7 col-xs-6 p5\">"+syado+"</dd>",'')
    height_ave = re.search(r'(?<=標高差：</dt><dd class="col-sm-7 col-xs-6 p5">).+?(?=<)', html).group(0)
    info_sabu.append(height_ave)
    index = html.find("</dt><dd class=\"col-sm-7 col-xs-6 p5\">"+height_ave+"</dd>")
    html = html[index:]
    html = html.replace("</dt><dd class=\"col-sm-7 col-xs-6 p5\">"+height_ave+"</dd>",'')
    long = re.search(r'(?<=最長滑走距離：</dt><dd class="col-sm-7 col-xs-6 p5">).+?(?=<)', html).group(0)
    info_sabu.append(long)
    try:
      revel_easy = re.search(r'(?<=初級<br/>).+?(?=<)', html).group(0)
    except:
      revel_easy = "10%"
    revel_normal = re.search(r'(?<=中級<br/>).+?(?=<)', html).group(0)
    revel_difficult = re.search(r'(?<=上級<br/>).+?(?=<)', html).group(0)
    info_sabu.append(revel_easy)
    info_sabu.append(revel_normal)
    info_sabu.append(revel_difficult)
    assetu = re.search(r'(?<=圧雪<br/>).+?(?=<)', html).group(0)
    try:
      hi_assetu = re.search(r'(?<=非圧雪<br/>).+?(?=<)', html).group(0)
    except:
      hi_assetu = "0%"
    try:
      kobu = re.search(r'(?<=コブ<br/>).+?(?=<)', html).group(0)
    except:
      kobu = "0%"
    info_sabu.append(assetu)
    info_sabu.append(hi_assetu)
    info_sabu.append(kobu)
    skiier = re.search(r'(?<=スキーヤー<br/>).+?(?=<)', html).group(0)
    boder = re.search(r'(?<=ボーダー<br/>).+?(?=<)', html).group(0)
    info_sabu.append(skiier)
    info_sabu.append(boder)
    info.append(info_sabu)
    print(info_sabu)
  except:
    continue


```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(info, columns=["最大高度", "最大斜度", "標高差", "平均高度", "初心者", "中級者", "上級者", "圧雪", "非圧雪", "コブ", "スキーヤー", "ボーダー"])
```
