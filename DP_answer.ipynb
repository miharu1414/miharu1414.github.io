```python

import pandas as pd

df = pd.read_excel("./input/data_num.xlsx") 

# 難易度の閾値
biginner = 40
middle = 40
expert = 30
diff = [biginner,middle,expert]

# スキーor スノボ
bias = 35
# 非圧雪多めの指標
hiassetsu = 25
# 滑走距離
dist = 3000
# トップ標高
top = 1500

# 条件の候補数
ans_num = [2,3,2,2,3]

```

```python
df.head()
```

```python
# いらない条件部分の削除
df = df.drop(['リフト数','コース数','最大斜度','標高差','コブ'], axis=1)
# columns の並び替え
df = df.reindex(columns=['スキー場名','url','スキーヤー','ボーダー','初級','中級','上級','非圧雪','最長滑走距離','トップ標高'])
df.head()
```

```python
list_data = []
loss = []
for i in range(len(df)):
    list = df.iloc[i].to_list()
    loss.append(0)
    list_data.append(list)
list_data
```

```python
def search(list_data,example,diff,bias,hiassetsu,dist,top):
    for example in examples:
        loss = []
        for gelende in range(len(list_data)):
            loss.append(0)
            for i in range(len(example)):

                # 気にしないの判定
                if i == 2|i==3:
                    if example[i] == '1':
                        continue
                if i == 4:
                    if example[i] == '2':
                        continue
            
                # スキーor ボード
                if i == 0:
                    if example[i] == '0':
                        if list_data[gelende][2] < bias:
                            loss[gelende]+=1
                    else:
                        if list_data[gelende][3] < bias:
                            loss[gelende]+=1
                            
                # 初級　中級　上級　の判定
                if i == 1:
                    level = int(example[i])
                    if list_data[gelende][4 + level] < diff[level]:
                        loss[gelende]+=1
                
                # 非圧雪に関して
                if i == 2:
                    if example[i] == '0':
                        if list_data[gelende][7] < hiassetsu:
                            loss[gelende]+=1
                            
                # 最長滑走距離
                if i == 3:
                    if example[i] == '0':
                        if list_data[gelende][8] < dist:
                            loss[gelende]+=1
                            
                # 標高に関して
                if i == 4:
                    if example[i] == '0':
                        if list_data[gelende][9] >= top:
                            loss[gelende]+=1
                    
                    elif example[i] == '1':
                        if list_data[gelende][9] < top:
                            loss[gelende]+=1
        index = loss.index(min(loss))
        print("'"+example+"'" +":"+str(index)+"," +  "  最適: " + str(list_data[index][0] + " url = " + str(list_data[index][1])))
```

```python
import copy
# 候補の作成 
examples =[""]
for i in range(len(ans_num)):
    examples_sb = []
    for j in range(ans_num[i]):
        for string in examples:
            ans = string + str(j)
            examples_sb.append(ans)
    examples = examples_sb
examples = examples_sb
print(examples)
print("候補数:"+str(len(examples)))
```

```python
search(list_data,examples,diff,bias,hiassetsu,dist,top)
```
